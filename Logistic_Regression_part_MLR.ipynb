```python
import numpy as np
from numpy.linalg import inv
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
pd.DataFrame(data["data"], data["target"])
```

```python
def sigmoid(y):
  a = (1/(1+np.exp(-y)))
  if a<0.5:
    return 0
  return 1
```

```python
X = data["data"]
X = np.hstack((np.ones((X.shape[0], 1)), X))
Y = np.reshape(data["target"], (569, 1))

print(f'X Shape: {X.shape}')
print(f'Y Shape: {Y.shape}')
```

```python
A = inv(np.matmul(X.T, X))
g = np.matmul(X.T, Y)

B = np.matmul(A, g)
```

```python
pd.DataFrame(B,columns=["B Vals"])
```

```python
def predict_y(x_param, B):
  y = 0
  for i in range(len(x_param)-1):
    y += (x_param[i] * B[i])

  return sigmoid(y)
```

```python
acc = 0

for i, j in zip(X, Y):
  if predict_y(i, B) == j:
    acc += 1

acc = acc/len(X)

print(f"Accuracy: {acc}")
```
